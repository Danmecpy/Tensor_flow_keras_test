```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import os
df = pd.read_excel(r'C:\Users\danme\OneDrive\Documentos\GitHub\Tensor_flow_keras_test\Auto.ipynb\Notas_3_trimestre.xlsx')
print(df.head())
class ExcelEditorApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Editor de Observações")
        
        # Caminho específico do arquivo
        self.filepath = r'C:\Users\danme\OneDrive\Documentos\GitHub\Tensor_flow_keras_test\Auto.ipynb\Notas_3_trimestre.xlsx'
        
        # Inicializa df como None
        self.df = None
        
        # Tenta carregar o arquivo Excel primeiro
        if not self.load_excel_file():
            return  # Se falhar ao carregar o arquivo, não continua a inicialização
        
        # Se chegou aqui, configura a interface
        self.setup_interface()

    def load_excel_file(self):
        """Carrega o arquivo Excel com as configurações específicas"""
        try:
            # Carrega o arquivo com header na linha 1
            self.df = pd.read_excel(self.filepath, header=1)
            
            # Adiciona a coluna 'ID' se não existir
            if 'ID' not in self.df.columns:
                self.df.insert(0, 'ID', range(1, len(self.df) + 1))

            # Verifica se a coluna 'Observações' existe
            if 'Observações' not in self.df.columns:
                messagebox.showerror("Erro", "A coluna 'Observações' deve estar presente na planilha.")
                return False

            return True  # Retorna True se tudo deu certo

        except FileNotFoundError:
            messagebox.showerror("Erro", f"Arquivo não encontrado: {self.filepath}")
            return False
        except Exception as e:
            messagebox.showerror("Erro", f"Erro ao carregar o arquivo: {str(e)}")
            return False

    def setup_interface(self):
        """Configura a interface gráfica"""
        # Frame principal para organizar os elementos
        main_frame = tk.Frame(self.root)
        main_frame.pack(padx=10, pady=10)

        # ID
        id_frame = tk.Frame(main_frame)
        id_frame.pack(fill=tk.X, pady=5)
        
        tk.Label(id_frame, text="ID:").pack(side=tk.LEFT)
        self.id_entry = tk.Entry(id_frame)
        self.id_entry.pack(side=tk.LEFT, padx=5)
        tk.Button(id_frame, text="Buscar Observação", 
                 command=self.buscar_observacao).pack(side=tk.LEFT)

        # Área de observações
        obs_frame = tk.Frame(main_frame)
        obs_frame.pack(fill=tk.BOTH, expand=True, pady=5)
        
        tk.Label(obs_frame, text="Observações:").pack(anchor=tk.W)
        self.text_observacao = tk.Text(obs_frame, height=5, width=40)
        self.text_observacao.pack(fill=tk.BOTH, expand=True)

        # Email
        email_frame = tk.Frame(main_frame)
        email_frame.pack(fill=tk.X, pady=5)
        
        tk.Label(email_frame, text="Email:").pack(side=tk.LEFT)
        self.entry_email = tk.Entry(email_frame, width=40)
        self.entry_email.pack(side=tk.LEFT, padx=5)

        # Botões
        button_frame = tk.Frame(main_frame)
        button_frame.pack(fill=tk.X, pady=5)
        
        tk.Button(button_frame, text="Salvar Alterações", 
                 command=self.salvar_observacao).pack(side=tk.LEFT, padx=5)
        tk.Button(button_frame, text="Enviar por Email", 
                 command=self.enviar_observacao_email).pack(side=tk.LEFT)

    def buscar_observacao(self):
        """Busca a observação baseada no ID"""
        try:
            id_valor = int(self.id_entry.get())
            observacao = self.df.loc[self.df['ID'] == id_valor, 'Observações'].values
            if observacao.size > 0:
                self.text_observacao.delete(1.0, tk.END)
                self.text_observacao.insert(tk.END, observacao[0])
            else:
                messagebox.showwarning("Aviso", "ID não encontrado.")
        except ValueError:
            messagebox.showerror("Erro", "Por favor, insira um ID válido.")

    def salvar_observacao(self):
        """Salva a observação editada"""
        try:
            id_valor = int(self.id_entry.get())
            nova_observacao = self.text_observacao.get(1.0, tk.END).strip()
            if self.df.loc[self.df['ID'] == id_valor, 'Observações'].size > 0:
                self.df.loc[self.df['ID'] == id_valor, 'Observações'] = nova_observacao
                try:
                    with pd.ExcelWriter(self.filepath, engine='openpyxl', mode='w') as writer:
                        self.df.to_excel(writer, index=False)
                    messagebox.showinfo("Sucesso", "Observação atualizada com sucesso!")
                except Exception as e:
                    messagebox.showerror("Erro", f"Falha ao salvar o arquivo: {e}")
            else:
                messagebox.showwarning("Aviso", "ID não encontrado.")
        except ValueError:
            messagebox.showerror("Erro", "Por favor, insira um ID válido.")

    def enviar_observacao_email(self):
        """Envia a observação por email"""
        try:
            id_valor = int(self.id_entry.get())
            observacao = self.text_observacao.get(1.0, tk.END).strip()
            if observacao:
                destinatario = self.entry_email.get().strip()
                if destinatario:
                    assunto = f"Observação Atualizada para ID {id_valor}"
                    mensagem = f"Prezado(a),\n\nA observação para o ID {id_valor} foi atualizada para:\n\n{observacao}\n\nAtenciosamente."
                    self.enviar_email(destinatario, assunto, mensagem)
                else:
                    messagebox.showwarning("Aviso", "Por favor, insira um e-mail de destinatário.")
            else:
                messagebox.showwarning("Aviso", "A observação está vazia.")
        except ValueError:
            messagebox.showerror("Erro", "Por favor, insira um ID válido.")

    def enviar_email(self, destinatario, assunto, mensagem):
        """Envia o email usando as configurações SMTP"""
        try:
            # Configurações do servidor SMTP
            smtp_server = 'smtp.seuprovedor.com'
            smtp_port = 587
            email_remetente = 'danfdts10@gmail.com'
            senha =  'geif cejg qjkj faez'

            msg = MIMEText(mensagem)
            msg['Subject'] = assunto
            msg['From'] = email_remetente
            msg['To'] = destinatario

            with smtplib.SMTP(smtp_server, smtp_port) as server:
                server.starttls()
                server.login(email_remetente, senha)
                server.sendmail(email_remetente, destinatario, msg.as_string())
            messagebox.showinfo("Sucesso", "E-mail enviado com sucesso!")
        except Exception as e:
            messagebox.showerror("Erro", f"Falha ao enviar e-mail: {e}")

def main():
    root = tk.Tk()
    app = ExcelEditorApp(root)
    # Só executa mainloop se o DataFrame foi carregado com sucesso
    if app.df is not None:
        root.mainloop()
    else:
        root.destroy()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import os

class ExcelEditorApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Editor de Observações")
        
        # Caminho específico do arquivo
        self.filepath = r'C:\Users\danme\OneDrive\Documentos\GitHub\Tensor_flow_keras_test\Auto.ipynb\Notas_3_trimestre.xlsx'
        
        # Inicializa df como None
        self.df = None
        
        # Tenta carregar o arquivo Excel primeiro
        if not self.load_excel_file():
            return
        
        # Se chegou aqui, configura a interface
        self.setup_interface()

    def load_excel_file(self):
        """Carrega o arquivo Excel com as configurações específicas"""
        try:
            # Tenta primeiro ler sem especificar header
            self.df = pd.read_excel(self.filepath)
            
            # Mostra as colunas disponíveis para debug
            print("Colunas disponíveis:", list(self.df.columns))
            
            # Procura pela coluna 'Observações' (case insensitive)
            obs_column = None
            for col in self.df.columns:
                if str(col).lower() == 'observações' or str(col).lower() == 'observacoes':
                    obs_column = col
                    break
            
            if obs_column:
                # Renomeia a coluna para ter o nome correto
                self.df = self.df.rename(columns={obs_column: 'Observações'})
            else:
                messagebox.showerror("Erro", f"Coluna 'Observações' não encontrada.\nColunas disponíveis: {list(self.df.columns)}")
                return False

            # Adiciona a coluna 'ID' se não existir
            if 'ID' not in self.df.columns:
                self.df.insert(0, 'ID', range(1, len(self.df) + 1))

            return True

        except FileNotFoundError:
            messagebox.showerror("Erro", f"Arquivo não encontrado: {self.filepath}")
            return False
        except Exception as e:
            messagebox.showerror("Erro", f"Erro ao carregar o arquivo: {str(e)}")
            return False

    def setup_interface(self):
        """Configura a interface gráfica"""
        # Frame principal para organizar os elementos
        main_frame = tk.Frame(self.root)
        main_frame.pack(padx=10, pady=10)

        # ID
        id_frame = tk.Frame(main_frame)
        id_frame.pack(fill=tk.X, pady=5)
        
        tk.Label(id_frame, text="ID:").pack(side=tk.LEFT)
        self.id_entry = tk.Entry(id_frame)
        self.id_entry.pack(side=tk.LEFT, padx=5)
        tk.Button(id_frame, text="Buscar Observação", 
                 command=self.buscar_observacao).pack(side=tk.LEFT)

        # Área de observações
        obs_frame = tk.Frame(main_frame)
        obs_frame.pack(fill=tk.BOTH, expand=True, pady=5)
        
        tk.Label(obs_frame, text="Observações:").pack(anchor=tk.W)
        self.text_observacao = tk.Text(obs_frame, height=5, width=40)
        self.text_observacao.pack(fill=tk.BOTH, expand=True)

        # Email
        email_frame = tk.Frame(main_frame)
        email_frame.pack(fill=tk.X, pady=5)
        
        tk.Label(email_frame, text="Email:").pack(side=tk.LEFT)
        self.entry_email = tk.Entry(email_frame, width=40)
        self.entry_email.pack(side=tk.LEFT, padx=5)

        # Botões
        button_frame = tk.Frame(main_frame)
        button_frame.pack(fill=tk.X, pady=5)
        
        tk.Button(button_frame, text="Salvar Alterações", 
                 command=self.salvar_observacao).pack(side=tk.LEFT, padx=5)
        tk.Button(button_frame, text="Enviar por Email", 
                 command=self.enviar_observacao_email).pack(side=tk.LEFT)

    def buscar_observacao(self):
        """Busca a observação baseada no ID"""
        try:
            id_valor = int(self.id_entry.get())
            observacao = self.df.loc[self.df['ID'] == id_valor, 'Observações'].values
            if observacao.size > 0:
                self.text_observacao.delete(1.0, tk.END)
                self.text_observacao.insert(tk.END, observacao[0])
            else:
                messagebox.showwarning("Aviso", "ID não encontrado.")
        except ValueError:
            messagebox.showerror("Erro", "Por favor, insira um ID válido.")

    def salvar_observacao(self):
        """Salva a observação editada"""
        try:
            id_valor = int(self.id_entry.get())
            nova_observacao = self.text_observacao.get(1.0, tk.END).strip()
            if self.df.loc[self.df['ID'] == id_valor, 'Observações'].size > 0:
                self.df.loc[self.df['ID'] == id_valor, 'Observações'] = nova_observacao
                try:
                    with pd.ExcelWriter(self.filepath, engine='openpyxl', mode='w') as writer:
                        self.df.to_excel(writer, index=False)
                    messagebox.showinfo("Sucesso", "Observação atualizada com sucesso!")
                except Exception as e:
                    messagebox.showerror("Erro", f"Falha ao salvar o arquivo: {e}")
            else:
                messagebox.showwarning("Aviso", "ID não encontrado.")
        except ValueError:
            messagebox.showerror("Erro", "Por favor, insira um ID válido.")

    def enviar_observacao_email(self):
        """Envia a observação por email"""
        try:
            id_valor = int(self.id_entry.get())
            observacao = self.text_observacao.get(1.0, tk.END).strip()
            if observacao:
                destinatario = self.entry_email.get().strip()
                if destinatario:
                    assunto = f"Observação Atualizada para ID {id_valor}"
                    mensagem = f"Prezado(a),\n\nA observação para o ID {id_valor} foi atualizada para:\n\n{observacao}\n\nAtenciosamente."
                    self.enviar_email(destinatario, assunto, mensagem)
                else:
                    messagebox.showwarning("Aviso", "Por favor, insira um e-mail de destinatário.")
            else:
                messagebox.showwarning("Aviso", "A observação está vazia.")
        except ValueError:
            messagebox.showerror("Erro", "Por favor, insira um ID válido.")
                
    def enviar_email(self, destinatario, assunto, mensagem):
        """Envia o email usando as configurações SMTP"""
        try:
            # Configurações do servidor SMTP para Gmail
            smtp_server = 'smtp.gmail.com'
            smtp_port = 587
            email_remetente = 'danfdts10@gmail.com'
            senha = 'geif cejg qjkj faez'

            msg = MIMEText(mensagem)
            msg['Subject'] = assunto
            msg['From'] = email_remetente
            msg['To'] = destinatario

            with smtplib.SMTP(smtp_server, smtp_port) as server:
                server.starttls()
                server.login(email_remetente, senha)
                server.sendmail(email_remetente, destinatario, msg.as_string())
            messagebox.showinfo("Sucesso", "E-mail enviado com sucesso!")
        except smtplib.SMTPException as e:
            messagebox.showerror("Erro", f"Falha ao enviar e-mail: {e}")
        except Exception as e:
            messagebox.showerror("Erro", f"Ocorreu um erro: {e}")

def main():
    root = tk.Tk()
    app = ExcelEditorApp(root)
    if app.df is not None:
        root.mainloop()
    else:
        root.destroy()

if __name__ == "__main__":
    main()
```
